```python
df['Company Name']= df['Company Name'].map(lambda x: x.split('\n')[0])
```

```python
df['Company Name']
```

```python
df.columns
```

```python
categorical_cols = ['Job Title', 'Salary Estimate', 'Job Description', 'Rating',
       'Company Name', 'Location', 'Headquarters', 'Size', 'Founded',
       'Type of ownership', 'Industry', 'Sector', 'Revenue', 'Competitors']
df[categorical_cols] = df[categorical_cols].replace('-1','na')
df[categorical_cols] = df[categorical_cols].replace('unknown','na')
df['Rating'] = df['Rating'].replace(-1.,0)
```

```python
df
```

```python
df['Founded'] = df['Founded'].replace(-1,0)
```

```python
df
```

```python
pattern = r'\$(\d+)K-\$(\d+)K'
df[['min_salary','max_salary']]=df['Salary Estimate'].str.extract(pattern).astype(int)
```

```python
df['avg_salary'] = df.apply(lambda row: int((row['min_salary']+row['max_salary'])/2),axis=1)
```

```python
df['Salary Estimate']=df['min_salary'].astype(str)+'-'+df['max_salary'].astype(str)
```

```python
df
```

```python
df['company_age'] = df['Founded'].map(lambda x: 2024 - x if x > 0 else 0)
```

```python
df['same_state']= df.apply(lambda row: 1 if row['Location']==row['Headquarters'] else 0,axis=1)
```

```python
df['python']= df['Job Description'].map(lambda x : 1 if 'python' in x.lower() else 0 )
df['excel'] = df['Job Description'].map(lambda x : 1 if 'excel' in x.lower() else 0)
df['hadoop'] = df['Job Description'].map(lambda x : 1 if 'hadoop' in x.lower() else 0)
df['aws'] = df['Job Description'].map(lambda x : 1 if 'aws' in x.lower() else 0)
df['spark'] = df['Job Description'].map(lambda x : 1 if 'spark' in x.lower() else 0)
df['big_data'] = df['Job Description'].map(lambda x : 1 if 'big_data' in x.lower() else 0)
```

```python
df
```

```python
df['Type of ownership']
```

```python
df.rename(columns={"Type of ownership":"type_of_ownership"}, inplace=True)
```

```python
def ownership(jobs):
    return (
        jobs.loc[:, "type_of_ownership"]
        .replace("Nonprofit Organization","Nonprofit")
        .replace("-1","n/a")
        .replace("Unknown","n/a")
        .replace("Company - Public","Public")
        .replace("Company - Private","Private")
        .replace("Other Organization","Other")   
    )
df = df.assign(type_of_ownership = ownership)
```

```python
def role_defining(title): 
    if any(keyword in title.lower() for keyword in ['data scientist', 'data science']):
        return 'Data Scientist'
    elif any(keyword in title.lower() for keyword in ['data analyst', 'analyst']):
        return 'Data Analyst'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif any(keyword in title.lower() for keyword in ['machine learning', 'ai']):
        return 'Machine Learning Engineer'
    else:
        return 'Other'
```

```python
df['Job Title'].value_counts()
```

```python
df['Job_role'] = df['Job Title'].apply(role_defining)
```

```python
df['Job_role'].value_counts()
```

```python
df['Job Title'][df['Job_role']=='other']
```

```python
def seniority(title, description):
    if any(keyword in title.lower() for keyword in  ["jr.",'jr','junior']) or any(keyword in description.lower() for keyword in ['jr.','jr','junior']):
        return 'junior'
    elif any(keyword in title.lower() for keyword in ['sr.','sr','senior']) or any(keyword in description.lower() for keyword in ['sr.','sr','senior']):
        return ' senior'
    else:
        return 'mid level'
```

```python
df['job_seniority'] = df.apply(lambda row: seniority(row['Job Title'], row['Job Description']),axis=1)
```

```python
df['job_seniority'].value_counts()
```
